```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

random_value = np.random.rand(100)

series = pd.Series(random_value)

df = pd.DataFrame(series, columns=['values'])

statics = df.describe()

statics.to_csv('statistacs.csv')

plt.plot(df.index, df['values'])

plt.title('Random Values')
plt.xlabel('Index')
plt.ylabel('Value')

plt.savefig('random_values_plot.png')

plt.show()
```
